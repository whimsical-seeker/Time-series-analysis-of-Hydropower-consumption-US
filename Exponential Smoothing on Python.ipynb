```python
import numpy as np
from scipy.optimize import minimize

# Define the time series data
data = np.array([10, 12, 14, 15, 18])
```

```python
def simple_exponential_smoothing(alpha, data):
    # Initialize the level (initial forecast) with the first data point
    F1 = data[0]
    sse = 0

    # Calculate forecasts, errors, and SSE
    for t in range(1, len(data)):
        Ft = alpha * data[t] + (1 - alpha) * F1
        et = data[t] - Ft
        sse += et**2

        # Update the level for the next iteration
        F1 = Ft

    return sse
```
